```python
import math
import numpy as np

import matplotlib.pyplot as plt

x = np.arange(-3., 3., 0.05)

def graph(x, y, name):
    
    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label=name, color='blue', linewidth=2)
    plt.axhline(y=0, color='black', linewidth=0.8)
    plt.axvline(x=0, color='black', linewidth=0.8)

    # Add grid lines
    plt.grid(True, linestyle='--', alpha=0.6)

    # Add labels and title
    plt.xlabel('x', fontsize=14)
    plt.ylabel('f(x)', fontsize=14)
    plt.legend(loc='upper left')

    # Customize the ticks
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)

    # Show the plot
    plt.show()
```

<h3 style="text-align: center;">Sigmaoid Function</h3>
<img src="https://wikimedia.org/api/rest_v1/media/math/render/svg/e5b42d3fad3c41825a1493a0daa271523cbab01c" alt="html5" />

```python
# Sigmaoid Function

def sigmaoid(x):
    
    a= []
    for i in x:
        a.append(1/(1+math.exp(-i)))
    return a
```

```python
graph(x, sigmaoid(x), 'Sigmaoid')
```

<h3 style="text-align: center;">Hyperbolic Tangent</h3>

<img src="https://wikimedia.org/api/rest_v1/media/math/render/svg/b8dc4c309a551cafc2ce5c883c924ecd87664b0f" alt="html5" />


```python
# Hyperbolic Tangent

def tanh(x, derivative = False):
    
    if (derivative == True):
        return (1-(x**2))
    return np.tanh(x)
```

```python
graph(x, tanh(x), 'Hyperbolic Tangent')
```

<h3 style="text-align: center;">ReLU (rectified linear unit)</h3>
<img src="https://wikimedia.org/api/rest_v1/media/math/render/svg/1b43bb9d7f851aa7c8ce8e4dacbd943d18512528" alt="html5" />

```python
# ReLU

def relu(x):
    b = []
    for i in x:
        if i < 0:
            b.append(0)
        else:
            b.append(i)
    return b
```

```python
graph(x, relu(x), 'ReLU')
```

<h3 style="text-align: center;">Leaky ReLU</h3>

<img src="https://wikimedia.org/api/rest_v1/media/math/render/svg/61643f7eb391b87efb76400adfff9f204a66fdb6" alt="html5" />

```python
# Leaky ReLU

def leaky_relu(x):
    b = []
    for i in x:
        if i < 0:
            b.append(i/10)
        else:
            b.append(i)
    return b
```

```python
graph(x, leaky_relu(x), 'Leaky ReLU')
```
